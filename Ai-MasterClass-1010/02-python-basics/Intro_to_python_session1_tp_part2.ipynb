# Functions

**A function is a reusable block of code.**

```python
def func1(n):
  a = n*2
  b = a+4
  c = b*12
  return c
```

```python
print(func1(3))
```

```python
def square(n):
  m = n**2
  return m
```

```python
def square2(n,m):
  return n**m
```

```python
print(square(3))
```

```python
print(square2(3,3))
```

# Classes

**Class**: temple for objects.

**Object**: Instance of a class.


**Difference**:

*   Function → standalone code.
*   Class → groups data (attributes) & methods.
*   Object (or Instance)  →  *An object is a specific* instance created from a class.
*   self: Inside a class's methods, self refers to the specific object that the method is being called on. It's how Python knows which car's details to display when you call my_car.display_info(). self is always the first parameter of any method within a class.
*   __init__ Method (The Constructor) → This special method is called automatically whenever you create a new object from a class. Its job is to "initialize" the object, usually by setting the initial values for its attributes.
*   Attributes →These are variables that belong to an object. They store the state or data for that specific object (e.g., self.name = "Buddy").
*   Methods → These are functions that belong to a class. They define the behaviors or actions an object can perform (e.g., def bark(self):).


**e.g** : If Car is the template, then your specific car (a 2022 Ford Mustang) and my car (a 2023 Honda Civic) are two separate objects created from that template. Each object has its own data (attributes) but shares the same behaviors (methods) defined by the class.








```python
class Car:
  def __init__(self,brand,model , year , price) :
    self.brand = brand
    self.model = model
    self.year = year
    self.price = price
  def display_info(self) :
    print(f"This is a {self.brand} {self.year} {self.model} {self.price}")

golf = Car("volkswagen","7 r", 2017 , 24000000000)
golf.display_info()
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name

        if isinstance(marks, int) or isinstance(marks, float):
            self.marks = [marks]  # convert single number to list
        else:
            self.marks = marks

    def average(self):
       return sum(self.marks) / len(self.marks)

    def status(self):
        if self.average() >= 10:
            return "accepted"
        else:
            return "rejected"
```

```python
sara = Student("Sara", [15, 16, 14])
ali = Student("Ali", [8, 9, 7, 10])
```

```python
print(sara.name, sara.marks, sara.average(), sara.status())
print(ali.name, ali.marks, ali.average(), ali.status())
```
