**Lists**

**List — ordered, mutable, allows duplicates**

**When to use:**

* You need to keep items in the order they were added.
* You might have duplicates.
* You need to frequently add/remove/change elements.

**Real app examples:**

* **E-commerce cart**: `[ "shirt", "shoes", "shirt" ]` → items can be repeated and ordered by when added.


---
 **Set — unordered, mutable, unique elements**

**When to use:**

* You only care about unique values.
* Order does not matter.
* You need fast lookups (“is this thing in the set?”).

**Real app examples:**

* **Tag system in a blog**: `{ "python", "ai", "python" }` → duplicates auto-removed.


---
**Tuple — ordered, immutable**

**When to use:**

* You need a fixed collection of values that shouldn’t change.
* Often used for “records” or “coordinates” that stay constant.

**Real app examples:**

* **GPS coordinates**: `(40.7128, -74.0060)` → latitude & longitude.
* **Settings constants**: days of week → `("Mon", "Tue", "Wed", ...)`.

**Exercise**

You are building a student registration system for a small course.

*   Use a list called student_list.
*   Add some names with append() and insert().
*   Remove one student by index.

Track all unique courses offered
*   Use a set called course_set.
*   Add new courses "math" ," art ","history" using add() function
*   Using update() add courses "math" , "science" , "AI".

```python
# 1️⃣ LIST – order matters, duplicates allowed, mutable
print("=== LIST ===")
student_list = []  # start empty

# Add students
student_list.append("AMINE")
student_list.append("MOHAMED")
student_list.insert(1, "DOUNIA")  # Insert Charlie at index 1
print("After adding:", student_list)

# Remove student by index
removed_student = student_list.pop(0)  # Remove Alice
print(f"Removed {removed_student}")
print("After removal:", student_list)


# 2️⃣ SET – unordered, unique elements, mutable
print("=== SET ===")
course_set = set()

# Add courses
course_set.add("Math")
course_set.add("Science")
course_set.update(["History", "Art", "AI" , "Math"])  # "Math" ignored (duplicate)
print("After adding:", course_set)

# Remove course by name
course_set.remove("Science")  # raises error if not found
print("After removal:", course_set)
```

# Conditions

**If - Else**

```python
a = 5
b = 10
if a>b:
  print("A greater than B")
else:
  print("B greater than A")
```

```python
temperature = 25

if temperature > 30:
    print("It's a hot day!")
elif temperature > 20:
    print("It's a pleasant day.")
else:
    print("It's a cool day.")
```

# Loops

**For loops**

```python
for i in range(5):  # range(5) generates numbers from 0 to 4
    print(f"Iteration {i}")
```

```python
for i in range(1, 5, 2):
  print(i)
```

```python
for fruit in fruits:
  print(fruit)
```

```python
numbers = [3, 5, 7, 9, 11]

for num in numbers:
    print("Checking:", num)
    if num == 7:
        print("Found 7! Stopping the loop.")
        break
    print("Still searching...")

print("Loop finished.")
```

```python
nums = [0,2,1,6,3]

print(len(nums))
print(sum(nums))
print(max(nums))
print(min(nums))
```

**While**

```python
val1 = True
val2 = False
print(val1 and val2)
print(val1 or val2)
```

```python
print(not val1 and  val2)
print(not val1 or  val2)
```

```python
count = 2
while count < 5:
    print(f"Count is {count}")
    count += 1 # This is the same as count = count + 1
```

```python
animals = ["cat", "tiger", "lion", "dog"]
found = False
i=0
while i < len(animals) and not found:
    if animals[i] == "lion":
        found = True
    i = i + 1
if found: ##found == True
    print("lion found")
else:
    print("lion not found")
```
