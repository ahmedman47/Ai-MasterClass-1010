# Numpy


## What is NumPy?

NumPy is a Python library used to handle numerical data through **multidimensional arrays**, also called **tensors**.

---

## Why is NumPy important?

* Very fast thanks to low-level optimized implementations
* More efficient than standard Python lists for numerical computations
* Designed to work with large amounts of data, vectors, and matrices

---

## 🔹 The `ndarray` concept

The core structure in NumPy is the **ndarray**, which can have different dimensions:

* **0D**: a single data point
* **1D**: a vector
* **2D**: a matrix
* **nD**: a **tensor** used in scientific computing and deep learning

---

## 🔹 Element-wise operations

NumPy performs operations **element-wise**, meaning each operation is applied to every element of the array automatically.
This makes numerical computation:

* faster
* simpler
* more efficient

---

## 🔹 Common uses

* Numerical analysis
* Matrix manipulation
* Scientific computing
* Foundation for machine learning and deep learning

## Ndarrays vs Lists


**Python lists**


*   Can store any type of data (flexible but slow)
*   Memory is scattered
*   Operations require loops
*   Not designed for numerical/scientific computing

**NumPy arrays (ndarray)**

*   Store elements of the same data type
*   Stored in contiguous memory blocks (like C arrays)
*   Operations are vectorized → no explicit loops
*   Optimized in C → much faster
*   Designed for math, matrices, statistics, images, machine learning

**Insert** and **Remove** new values

```python
import numpy as np

# 1D array
arr1 = np.array([1, 2, 3, 4])
print("Original 1D:", arr1)

# Insert value 99 at index 2
arr1_inserted = np.insert(arr1, 2, 99)
print("After insert:", arr1_inserted)

# Remove element at index 1
arr1_removed = np.delete(arr1_inserted, 1)
print("After remove:", arr1_removed)

```

```python
# 2D array
arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])
print("\nOriginal 2D:\n", arr2)

# Insert new row at index 1
new_row = [9, 9, 9]
arr2_inserted_row = np.insert(arr2, 1, new_row, axis=0)
print("After inserting row:\n", arr2_inserted_row)

# Remove row at index 0
arr2_removed_row = np.delete(arr2_inserted_row, 0, axis=0)
print("After removing row:\n", arr2_removed_row)

# Insert new column at index 2
new_col = [8, 8]
arr2_inserted_col = np.insert(arr2, 2, new_col, axis=1)
print("After inserting column:\n", arr2_inserted_col)

# Remove column at index 1
arr2_removed_col = np.delete(arr2_inserted_col, 1, axis=1)
print("After removing column:\n", arr2_removed_col)
```

1. count
Why? → Tells you how many actual values you have (ignores missing values).

Use: Helps you spot missing data quickly. If count < total rows → some values are NaN.

2. mean
Why? → Shows the average, the central “balance point” of your numbers.

Use: Good for getting a quick idea of the typical value, but sensitive to extreme values (outliers).

3. std (standard deviation)
Why? → Measures how spread out your data is around the mean.

Use:

Low std = values are close together.

High std = values are widely spread.

4. min
Why? → The smallest value in the column.

Use: Helps detect extreme low values or errors (like a negative age).

5. 25% (Q1)
Why? → The value below which 25% of your data falls.

Use:

Helps see the lower range of “normal” values.

Used in boxplots to find outliers.

6. 50% (median)
Why? → The middle value — 50% of values are below it, 50% are above it.

Use:

Resistant to outliers (unlike mean).

Shows a “typical” value in skewed data.

7. 75% (Q3)
Why? → The value below which 75% of your data falls.

Use:

Shows the upper range of “normal” values.

Also used in boxplots for outlier detection.

8. max
Why? → The largest value in the column.

Use: Helps detect extreme high values or possible errors.

plt.subplots(rows, cols) → creates multiple axes

figsize=(width, height) → figure size in inches

ax[i] → reference to each subplot



```python
fig, ax = plt.subplots(1, 3, figsize=(10, 4))

ax[0].plot(x, y, color='blue')
ax[0].set_title("Line Plot")

ax[1].bar(categories, values, color='green')
ax[1].set_title("Bar Plot")

ax[2].scatter(x, y, color='red')
ax[2].set_title("Scatter Plot")

plt.tight_layout() #Adjust the padding between and around subplots. To exclude an artist on the Axes from the bounding box calculation that determines the subplot parameters.

plt.show()

```
