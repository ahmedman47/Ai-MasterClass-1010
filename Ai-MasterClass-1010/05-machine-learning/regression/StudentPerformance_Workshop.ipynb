## 🧹 Step 5 — Check for Missing Values

Missing values (empty cells) can cause problems in analysis.  
We check for them using `df.isnull().sum()`.

````python
print('🔎 Missing values per column:')
print(df.isnull().sum())

total_missing = df.isnull().sum().sum()
print(f'\n✅ Total missing values: {total_missing}')
````

---
## ✏️ Step 6 — Rename Columns (Clean Names)

Column names with spaces can be hard to work with.  
Let's rename them to use underscores `_` instead of spaces.

````python
# Rename all columns: lowercase + replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')

print('✅ New column names:')
print(df.columns.tolist())
````

---
## ➕ Step 7 — Create New Columns

We can **create new columns** by calculating from existing ones.  

Let's create:
1. `total_score` → sum of all 3 scores
2. `average_score` → average of all 3 scores

````python
# Create a total score column
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']

# Create an average score column (rounded to 2 decimal places)
df['average_score'] = (df['total_score'] / 3).round(2)

print('✅ New columns added:')
df[['math_score', 'reading_score', 'writing_score', 'total_score', 'average_score']].head()
````

---
## 🏷️ Step 8 — Add a Grade Column

Let's classify each student with a **letter grade** based on their average score:

| Average Score | Grade |
|--------------|-------|
| 90 and above | A     |
| 80 – 89      | B     |
| 70 – 79      | C     |
| 60 – 69      | D     |
| Below 60     | F     |

````python
# Function to assign a grade
def assign_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

# Apply the function to each row
df['grade'] = df['average_score'].apply(assign_grade)

print('✅ Grade column added!')
df[['average_score', 'grade']].head(10)
````

---
## 🔎 Step 9 — Filtering Data

We can **filter** rows based on a condition — like searching in Excel.

### 9.1 — Students who passed (average ≥ 60)

````python
# Filter: students who passed
passed = df[df['average_score'] >= 60]

print(f'✅ Number of students who passed: {len(passed)}')
passed.head()
````

### 9.2 — Students who completed test preparation

````python
# Filter: students who completed test preparation
prepared = df[df['test_preparation_course'] == 'completed']

print(f'✅ Students who completed prep course: {len(prepared)}')
prepared.head()
````

---
## 📈 Step 10 — Grouping & Aggregation

`groupby()` lets us **group** data by a category and calculate statistics per group.

### 10.1 — Average scores by gender

````python
# Group by gender and calculate mean scores
gender_avg = df.groupby('gender')[['math_score', 'reading_score', 'writing_score', 'average_score']].mean().round(2)

print('📊 Average Scores by Gender:')
gender_avg
````

### 10.2 — Average scores by test preparation course

````python
prep_avg = df.groupby('test_preparation_course')[['math_score', 'reading_score', 'writing_score']].mean().round(2)

print('📊 Average Scores by Test Preparation:')
prep_avg
````

### 10.3 — Count of grades

````python
grade_counts = df['grade'].value_counts().sort_index()

print('📊 Number of students per grade:')
print(grade_counts)
````

---
## 📉 Step 11 — Sorting Data

`sort_values()` sorts the table by a column.

### Top 5 students by average score

````python
# Sort by average score, highest first
top_students = df.sort_values('average_score', ascending=False).head(5)

print('🏆 Top 5 Students:')
top_students[['gender', 'average_score', 'grade', 'test_preparation_course']]
````

````python
# Bottom 5 students
bottom_students = df.sort_values('average_score', ascending=True).head(5)

print('📉 Bottom 5 Students:')
bottom_students[['gender', 'average_score', 'grade', 'test_preparation_course']]
````

---
## 🎨 Step 12 — Visualizations (Charts)

Now let's **visualize** our data with charts!

### 12.1 — Distribution of Average Scores (Histogram)

````python
plt.figure(figsize=(8, 5))
plt.hist(df['average_score'], bins=20, color='steelblue', edgecolor='white')
plt.title('Distribution of Average Scores', fontsize=14)
plt.xlabel('Average Score')
plt.ylabel('Number of Students')
plt.axvline(df['average_score'].mean(), color='red', linestyle='--', label=f"Mean: {df['average_score'].mean():.1f}")
plt.legend()
plt.tight_layout()
plt.show()
````

### 12.2 — Average Scores by Gender (Bar Chart)

````python
gender_avg.plot(kind='bar', figsize=(8, 5), colormap='Set2', edgecolor='white')
plt.title('Average Scores by Gender', fontsize=14)
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
````

### 12.3 — Grade Distribution (Pie Chart)

````python
plt.figure(figsize=(6, 6))
grade_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140,
                  colors=['#2ecc71','#3498db','#f1c40f','#e67e22','#e74c3c'])
plt.title('Grade Distribution', fontsize=14)
plt.ylabel('')
plt.tight_layout()
plt.show()
````

### 12.4 — Math vs Reading Score (Scatter Plot)

````python
plt.figure(figsize=(8, 5))
colors = {'male': 'steelblue', 'female': 'salmon'}
for gender, group in df.groupby('gender'):
    plt.scatter(group['math_score'], group['reading_score'],
                label=gender, alpha=0.6, color=colors[gender])
plt.title('Math Score vs Reading Score', fontsize=14)
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.legend()
plt.tight_layout()
plt.show()
````

### 12.5 — Effect of Test Preparation on Scores (Boxplot)

````python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='test_preparation_course', y='average_score',
            palette='Set2')
plt.title('Average Score by Test Preparation Course', fontsize=14)
plt.xlabel('Test Preparation')
plt.ylabel('Average Score')
plt.tight_layout()
plt.show()
````
